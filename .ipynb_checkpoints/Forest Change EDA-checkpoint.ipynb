```python
def find_loss_by_region(district, df):
    '''return forest loss per year for a given region '''
    loss = {}
    for year in range(2001, 2020):
        loss_area = int(df.loc[df['subnational1'] == str(district), str(year)])
        loss[year] = loss_area
    
    return loss
```

```python
region = subnat1[['subnational1', 'area_ha', 'gain_2000-2012_ha', 'total_loss_01_12']]
reg = [reg for reg in subnat1_tl.subnational1.unique()]

def get_region_data(df, region_name):
    total_area = int(df.loc[(df['subnational1'] == region_name)].area_ha)
    area_loss_01_12 = int(df.loc[(df['subnational1'] == region_name)].total_loss_01_12)
    area_gain_00_12 = int(df.loc[(df['subnational1'] == region_name)]['gain_2000-2012_ha'])

    return [region_name, total_area, area_loss_01_12, area_gain_00_12]

print('{')
for region_name in reg:
    info_list = get_region_data(region, region_name)
    reg_data = {'total': info_list[1], 'loss01_12': info_list[2], 'gain_00_12': info_list[3]}
    print(f"'{info_list[0]}' : ", reg_data, ",")
print('}')
```

```python
# figuring out some "growth rate" so i could model something
# for country growth
c_g= tc_loss
# finding the mean forest area loss
c_g['growth'] = round(((c_g['2002'] - c_g['2001']) + (c_g['2003'] - c_g['2002']) + (c_g['2004'] - c_g['2003']) + (c_g['2005'] - c_g['2004']) + (c_g['2006'] - c_g['2005']) + (c_g['2007'] - c_g['2006']) + (c_g['2008'] - c_g['2007']) + (c_g['2009'] - c_g['2008']) + (c_g['2010'] - c_g['2009']) + (c_g['2011'] - c_g['2010']) + (c_g['2012'] - c_g['2011']) + (c_g['2013'] - c_g['2012']) + (c_g['2014'] - c_g['2013']) + (c_g['2015'] - c_g['2014']) + (c_g['2016'] - c_g['2015']) + (c_g['2017'] - c_g['2016']) + (c_g['2018'] - c_g['2017']) + (c_g['2019'] - c_g['2018'])) / 18, 2)
########################################
# getting the mean of forest area loss across all threshold
c_g_rate = round(c_g['growth'].mean(),2)
########################################
# getting forest cover area from 2000 and 2010
c_g['cover_2000'] = country_tl.extent_2000_ha
c_g['cover_2010'] = country_tl.extent_2010_ha
# finding the average forest cover of myanmar
c_g['cover_mean'] = (c_g['cover_2000'] + c_g['cover_2010']) / 2
# getting the mean of average forest cover across all threshold
c_g_fc_mean = round(c_g['cover_mean'].mean(), 2)
#########################################
# finding the average loss rate percentage
c_g_f_cover_perc = round((c_g_rate / c_g_fc_mean) * 100, 5)
#########################################

myanmar_total_land_area = country_tl.area_ha
myanmar_total_land_area = int(myanmar_total_land_area.mean())
myanmar_forest_cover = c_g_fc_mean
myanmar_forest_cover_perc = round((c_g_fc_mean / myanmar_total_land_area) * 100, 3)

print("MM loses an average of {}Ha of forests every year at the rate of {}%".format(c_g_rate, c_g_f_cover_perc))
print("{}% of myanmar's total land is covered with forests as of 2010".format(myanmar_forest_cover_perc))
```

## Simple analysis by townships within states

```python
# will focus on threshold at 30 only
subnat2_tl = subnat2_tl.loc[subnat2_tl['threshold'] == 30]

# this function will get you all the "townships" under a specific district
def get_twshp_under_region(df, region):
    twnshp = df.loc[(df['subnational1'] == region)]
    return twnshp

subnat2_tl
```

```python
for region_name in reg:
    # getting the region
    region = get_twshp_under_region(subnat2_tl, region_name)

    years = {}
    for coln in ['tc_loss_ha_2001',
           'tc_loss_ha_2002', 'tc_loss_ha_2003', 'tc_loss_ha_2004',
           'tc_loss_ha_2005', 'tc_loss_ha_2006', 'tc_loss_ha_2007',
           'tc_loss_ha_2008', 'tc_loss_ha_2009', 'tc_loss_ha_2010',
           'tc_loss_ha_2011', 'tc_loss_ha_2012', 'tc_loss_ha_2013',
           'tc_loss_ha_2014', 'tc_loss_ha_2015', 'tc_loss_ha_2016',
           'tc_loss_ha_2017', 'tc_loss_ha_2018', 'tc_loss_ha_2019']:
        years[coln] = coln.strip("tc_losha")

    region = region.rename(years, axis='columns')
    region = region[['subnational1', 'subnational2', 'area_ha', '2001', '2002','2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011','2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
    
    # getting subregions
    subregion_loss = {}
    for i in range(len(region[['subnational2']])):
        subregion = region[['subnational2']].iloc[i].values[0]
        #getting yearly loss of subregions
        year_dict = {}
        for year in range(2001, 2020):
            year_dict[year] = int(region.loc[(region['subnational2'] == subregion), str(year)])        
        subregion_loss[subregion] = year_dict
    
    print({region_name: subregion_loss})
    print()

    
```

```python
ayer = get_twshp_under_region(subnat2_tl, 'Ayeyarwady')

years = {}
for coln in ['tc_loss_ha_2001',
       'tc_loss_ha_2002', 'tc_loss_ha_2003', 'tc_loss_ha_2004',
       'tc_loss_ha_2005', 'tc_loss_ha_2006', 'tc_loss_ha_2007',
       'tc_loss_ha_2008', 'tc_loss_ha_2009', 'tc_loss_ha_2010',
       'tc_loss_ha_2011', 'tc_loss_ha_2012', 'tc_loss_ha_2013',
       'tc_loss_ha_2014', 'tc_loss_ha_2015', 'tc_loss_ha_2016',
       'tc_loss_ha_2017', 'tc_loss_ha_2018', 'tc_loss_ha_2019']:
    years[coln] = coln.strip("tc_losha")

ayer = ayer.rename(years, axis='columns')
ayer = ayer[['subnational1', 'subnational2', 'area_ha', '2001', '2002','2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011','2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]

df = ayer[['subnational2', '2019']].sort_values(by='2019', ascending=False)
df.set_index('subnational2', inplace=True)

df.to_json()
```

```python
for year in range(2001, 2020):
    plt.figure(figsize=(6,3))
    plt.title('Bar graph of Forest Cover Loss by Township for Year {}'.format(year))
    sns.barplot(x=ayer['subnational2'], y=ayer[str(year)])
    plt.xticks(rotation=75)
    plt.ylim(0,300)
    plt.xlabel('Township')
    plt.ylabel('Area Loss (Ha)')
```
